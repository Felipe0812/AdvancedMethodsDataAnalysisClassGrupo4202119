# 03.1 - Plot and analyze the ACF and PACF of the 4 time series in data_arma.csv:

### What do these plots suggest about the data being an AR or MA process?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
%matplotlib inline
sns.set()
```

```python
# Cargar dataset
data_arma_df = pd.read_csv('../datasets/data_arma.csv')
data_arma_df.head()
```

```python
plot_acf( data_arma_df.ts1, lags=50);
plot_pacf( data_arma_df.ts1, lags=50);
```

```python
plot_acf( data_arma_df.ts2, lags=50);
plot_pacf( data_arma_df.ts2, lags=50);
```

```python
plot_acf( data_arma_df.ts3, lags=50);
plot_pacf( data_arma_df.ts3, lags=50);
```

```python
plot_acf( data_arma_df.ts4, lags=50);
plot_pacf( data_arma_df.ts4, lags=50);
```
